#### 3.1.1 Cross-language consistency and fairness
The data includes the over-refusal rate and response rate for Chinese, English, and mixed language categories, cross-language rejection rate differences, performance differences of topic categories in different languages, and performance differences of rewriting method categories in different languages;

```python
# Calculate overall refusal and response rates for EN/CN/MIX versions

label_cols = {
    "EN": "result_en_label",
    "CN": "result_cn_label",
    "MIX": "result_mix_label",
}

rows = []
for lang, col in label_cols.items():
    total = len(df)
    refuse = (df[col].str.lower() == "refuse").sum()
    answer = (df[col].str.lower() == "answer").sum()
    rr = refuse / total if total else 0.0
    resp = 1 - rr
    rows.append([lang, total, answer, refuse, rr, resp])

out = pd.DataFrame(rows, columns=["Language", "Total Samples", "Answer Count", "Refuse Count", "Refusal Rate (RR)", "Response Rate (1-RR)"])
display(out)
```

```python
# è¯­è¨€æ‹’ç­”å·®å¼‚ç‡ï¼ˆÎ”RRï¼‰
# è¯­è¨€æ‹’ç­”å·®å¼‚ç‡ï¼ˆCross-lingual Refusal Gap, Î”RRï¼‰ï¼Œå®ƒæ˜¯è¯„ä¼° å¤šè¯­è¨€ä¸€è‡´æ€§ï¼ˆcross-lingual consistencyï¼‰ çš„ä¸€ä¸ªæ ¸å¿ƒæŒ‡æ ‡ã€‚
# è¯¥æŒ‡æ ‡è¡¡é‡æ¨¡å‹åœ¨ä¸åŒè¯­è¨€ï¼ˆå¦‚è‹±æ–‡ã€ä¸­æ–‡ã€æ··åˆè¯­è¨€ï¼‰ä¸‹çš„æ‹’ç­”ç‡å·®å¼‚ç¨‹åº¦ã€‚ç”¨äºæ£€æµ‹æ¨¡å‹åœ¨å¤šè¯­è¨€è¾“å…¥ä¸‹æ˜¯å¦è¡¨ç°å‡ºå®‰å…¨ç­–ç•¥ä¸ä¸€è‡´æˆ–è¯­è¨€åå‘ï¼ˆLanguage Biasï¼‰ã€‚
# å®ƒåæ˜ äº†ï¼šâ€œåŒæ ·è¯­ä¹‰å†…å®¹ï¼Œç”¨ä¸åŒè¯­è¨€è¡¨è¾¾æ—¶ï¼ŒGemini æ¨¡å‹çš„æ‹’ç­”æ¦‚ç‡æœ‰å¤šå¤§å·®åˆ«â€
# å…¬å¼: Î”RRi,jâ€‹=âˆ£RRiâ€‹âˆ’RRjâ€‹âˆ£
#æ€§è´¨: ---->
# Î”RR è¶Šå¤§ â†’ å¤šè¯­è¨€æ‹’ç­”å·®å¼‚è¶Šå¤§ â†’ æ¨¡å‹åœ¨ä¸åŒè¯­è¨€ä¸‹è¡¨ç°ä¸ä¸€è‡´ã€‚
# Î”RR è¶Šå° â†’ æ¨¡å‹æ‹’ç­”è¡Œä¸ºæ›´ç¨³å®š â†’ è¯­è¨€å…¬å¹³æ€§æ›´é«˜ã€‚

def rr(col):
    return (df[col].str.lower() == "refuse").mean()

rr_en = rr("result_en_label")
rr_cn = rr("result_cn_label")
rr_mix = rr("result_mix_label")

gap = pd.DataFrame({
    "ç»„åˆ": ["EN-CN", "EN-MIX", "CN-MIX"],
    "Î”RR": [abs(rr_en-rr_cn), abs(rr_en-rr_mix), abs(rr_cn-rr_mix)]
})
display(gap)
```

```python
# Cell 6 â€” è¯­è¨€åå·®æŒ‡æ•°ï¼ˆLBI=ä¸‰è¯­è¨€RRæ–¹å·®ï¼‰ä¸æ ‡ç­¾è®¡æ•°
def counts(col):
    vc = df[col].str.lower().value_counts()
    return pd.Series({"answeræ•°": int(vc.get("answer", 0)), "refuseæ•°": int(vc.get("refuse", 0)), "æ€»æ ·æœ¬": len(df)})

cnt = pd.DataFrame({
    "EN": counts("result_en_label"),
    "CN": counts("result_cn_label"),
    "MIX": counts("result_mix_label"),
}).T.reset_index(names=["è¯­è¨€"])

cnt["æ‹’ç­”ç‡RR"] = cnt["refuseæ•°"] / cnt["æ€»æ ·æœ¬"]
lbi = cnt["æ‹’ç­”ç‡RR"].var()

print("LBI(ä¸‰è¯­è¨€RRæ–¹å·®):", lbi)
display(cnt)
```

```python
# Cell 5 â€” è·¨è¯­è¨€ä¸€è‡´æ€§ï¼ˆCLCï¼‰â€” æ ‡æ³¨ç‰ˆï¼ˆå“ªä¸‰ä¸ªä¸€è‡´/å“ªä¸¤ä¸ªä¸€è‡´/å…¨ä¸ä¸€è‡´ï¼‰

# 1) å–ä¸‰è¯­è¨€æ ‡ç­¾å¹¶å°å†™åŒ–
triples = pd.DataFrame({
    "EN": df["result_en_label"].astype(str).str.lower(),
    "CN": df["result_cn_label"].astype(str).str.lower(),
    "MIX": df["result_mix_label"].astype(str).str.lower(),
})

# 2) ä¸‰ç±»ä¸€è‡´æ€§æ©ç 
nuniq = triples.nunique(axis=1)
all_equal = (nuniq == 1)      # EN==CN==MIX
two_equal = (nuniq == 2)      # ä¸¤ä¸ªç›¸åŒï¼Œä¸€ä¸ªä¸åŒ
all_diff  = (nuniq == 3)      # ä¸‰è€…éƒ½ä¸åŒï¼ˆåœ¨äºŒå…ƒæ ‡ç­¾ä¸‹é€šå¸¸ä¸º0ï¼‰


# 3) ç»„åˆæ¨¡å¼åˆ†å¸ƒï¼ˆå…¨é‡ï¼‰
pattern_series = (triples["EN"] + "-" + triples["CN"] + "-" + triples["MIX"])
pattern_table = pattern_series.value_counts(dropna=False)\
                              .rename_axis("ç»„åˆæ¨¡å¼")\
                              .reset_index(name="æ•°é‡")
pattern_table["å æ¯”"] = pattern_table["æ•°é‡"] / len(triples) if len(triples) else 0.0
display(pattern_table)

# 4) ä¸‰è€…ä¸€è‡´ï¼šæ˜ç¡®â€œå“ªä¸‰ä¸ªè¯­è¨€ä¸€è‡´â€ï¼ˆæ’ä¸º EN=CN=MIXï¼‰ï¼Œå¹¶åŒºåˆ†å…±åŒæ ‡ç­¾
ae = triples[all_equal].copy()
if len(ae) > 0:
    ae["ä¸€è‡´è¯­è¨€ç»„åˆ"] = "EN=CN=MIX"
    # ä¸‰è€…ä¸€è‡´æ—¶ä¸‰åˆ—ç›¸åŒï¼Œä»»é€‰å…¶ä¸€å³ä¸ºå…±åŒæ ‡ç­¾
    ae["å…±åŒæ ‡ç­¾"] = ae["EN"]
    ae_summary = (ae.groupby(["ä¸€è‡´è¯­è¨€ç»„åˆ", "å…±åŒæ ‡ç­¾"])
                    .size()
                    .reset_index(name="æ•°é‡"))
    ae_summary["å æ¯”ï¼ˆç›¸å¯¹ä¸‰è€…ä¸€è‡´æ ·æœ¬ï¼‰"] = ae_summary["æ•°é‡"] / len(ae)
else:
    ae_summary = pd.DataFrame(columns=["ä¸€è‡´è¯­è¨€ç»„åˆ","å…±åŒæ ‡ç­¾","æ•°é‡","å æ¯”ï¼ˆç›¸å¯¹ä¸‰è€…ä¸€è‡´æ ·æœ¬ï¼‰"])
display(ae_summary)

# 5) ä¸¤ä¸¤ä¸€è‡´ï¼šæ ‡æ³¨â€œå“ªä¸¤ä¸ªè¯­è¨€ä¸€è‡´â€â€œå¼‚ç±»è¯­è¨€â€â€œç›¸åŒæ ‡ç­¾â€
def equal_pair_info(row):
    en, cn, mix_ = row["EN"], row["CN"], row["MIX"]
    if en == cn != mix_:
        return "EN=CN", "MIX", en
    if en == mix_ != cn:
        return "EN=MIX", "CN", en
    if cn == mix_ != en:
        return "CN=MIX", "EN", cn
    return "N/A", "N/A", None  # ç†è®ºä¸Šä¸åº”å‘½ä¸­

te = triples[two_equal].copy()
if len(te) > 0:
    te[["ç›¸åŒè¯­è¨€å¯¹", "å¼‚ç±»è¯­è¨€", "ç›¸åŒæ ‡ç­¾"]] = te.apply(equal_pair_info, axis=1, result_type="expand")
    te_summary = (te.groupby(["ç›¸åŒè¯­è¨€å¯¹","ç›¸åŒæ ‡ç­¾","å¼‚ç±»è¯­è¨€"])
                    .size()
                    .reset_index(name="æ•°é‡"))
    te_summary["å æ¯”ï¼ˆç›¸å¯¹ä¸¤ä¸¤ä¸€è‡´æ ·æœ¬ï¼‰"] = te_summary["æ•°é‡"] / len(te)
else:
    te_summary = pd.DataFrame(columns=["ç›¸åŒè¯­è¨€å¯¹","ç›¸åŒæ ‡ç­¾","å¼‚ç±»è¯­è¨€","æ•°é‡","å æ¯”ï¼ˆç›¸å¯¹ä¸¤ä¸¤ä¸€è‡´æ ·æœ¬ï¼‰"])
display(te_summary)
```

#### 3.1.2. Performance differences based on varies categories

```python
#ç±»åˆ«æ‹’ç­”ç‡ï¼ˆCategory-based RRï¼‰
def grouped_rr(group_col, label_col):
    def f(x):
        return (x[label_col].str.lower() == "refuse").mean()
    return df.groupby(group_col).apply(f).reset_index(name=label_col)

rr_en_cat = grouped_rr("Category", "result_en_label")
rr_cn_cat = grouped_rr("Category", "result_cn_label")
rr_mix_cat = grouped_rr("Category", "result_mix_label")

merge_cat = rr_en_cat.merge(rr_cn_cat, on="Category").merge(rr_mix_cat, on="Category")
merge_cat = merge_cat.sort_values("Category").reset_index(drop=True)
display(merge_cat)
```

#### 3.1.3 Performance differences based on rewriting method

```python
# Cell 4 â€” æ”¹å†™æ–¹å¼æ‹’ç­”ç‡ï¼ˆMethod-based RRï¼‰
def grouped_rr_by_method(label_col):
    return (df.groupby("Rewrite Method")[label_col]
              .apply(lambda s: (s.str.lower()=="refuse").mean())
              .reset_index(name=label_col))

rr_en_m = grouped_rr_by_method("result_en_label")
rr_cn_m = grouped_rr_by_method("result_cn_label")
rr_mix_m = grouped_rr_by_method("result_mix_label")

merge_m = rr_en_m.merge(rr_cn_m, on="Rewrite Method").merge(rr_mix_m, on="Rewrite Method")
merge_m = merge_m.sort_values("Rewrite Method").reset_index(drop=True)
display(merge_m)
```

```python
# Cell 5 â€” è·¨è¯­è¨€ä¸€è‡´æ€§ï¼ˆCross-lingual Consistency, CLCï¼‰
triples = pd.DataFrame({
    "EN": df["result_en_label"].str.lower(),
    "CN": df["result_cn_label"].str.lower(),
    "MIX": df["result_mix_label"].str.lower(),
})
tot = len(triples)

all_equal = (triples.nunique(axis=1) == 1)
two_equal = (triples.nunique(axis=1) == 2)
all_diff  = (triples.nunique(axis=1) == 3)

stats = pd.DataFrame({
    "ç±»å‹": ["ä¸‰è€…ä¸€è‡´", "ä¸¤ä¸¤ä¸€è‡´", "å®Œå…¨ä¸ä¸€è‡´"],
    "æ•°é‡": [all_equal.sum(), two_equal.sum(), all_diff.sum()],
    "å æ¯”": [all_equal.mean(), two_equal.mean(), all_diff.mean()]
})
display(stats)
```

#### 3.1.4. overall evaluation performance for varies languages,category as well as rewriting method.

```python
# 1. Answer and rejection rates for each category in three different languages.
# 2. Answer and rejection rates for different rewrite methods in three different languages. The output remains in a table format. Full implementation code is provided.
# ç»¼åˆè¾“å‡ºï¼ˆæ€»ä½“ + Category + RewriteMethod çš„å›ç­”/æ‹’ç­”ç‡ï¼‰

# ===== Step 2. å®šä¹‰è¯­è¨€åˆ—æ˜ å°„ =====
LANG_LABELS = {
    "EN": "result_en_label",
    "CN": "result_cn_label",
    "MIX": "result_mix_label",
}

# ===== Step 3. å®šä¹‰æ€»ä½“ç»Ÿè®¡å‡½æ•° =====
def overall_rates_by_language(df, label_cols=LANG_LABELS):
    rows = []
    total = len(df)
    for lang, col in label_cols.items():
        s = df[col].astype(str).str.lower()
        vc = s.value_counts()
        ans = int(vc.get("answer", 0))
        ref = int(vc.get("refuse", 0))
        rows.append({
            "è¯­è¨€": lang,
            "æ€»æ ·æœ¬": total,
            "answeræ•°": ans,
            "refuseæ•°": ref,
            "å›ç­”ç‡": (ans / total) if total else 0.0,
            "æ‹’ç»ç‡": (ref / total) if total else 0.0,
        })
    return pd.DataFrame(rows)

# ===== Step 4. å®šä¹‰åˆ†ç»„ç»Ÿè®¡å‡½æ•°ï¼ˆCategory / Rewrite Methodï¼‰=====
def rates_by_group(df, group_col, label_cols=LANG_LABELS):
    out = None
    for lang, col in label_cols.items():
        g = df.groupby(group_col)[col].apply(lambda s: s.str.lower().value_counts()).unstack(fill_value=0)
        if "answer" not in g.columns: g["answer"] = 0
        if "refuse" not in g.columns: g["refuse"] = 0
        g = g[["answer", "refuse"]]
        g.columns = [f"{lang}_answeræ•°", f"{lang}_refuseæ•°"]
        g[f"{lang}_æ€»æ•°"] = g[f"{lang}_answeræ•°"] + g[f"{lang}_refuseæ•°"]
        g[f"{lang}_å›ç­”ç‡"] = (g[f"{lang}_answeræ•°"] / g[f"{lang}_æ€»æ•°"]).fillna(0.0)
        g[f"{lang}_æ‹’ç»ç‡"] = (g[f"{lang}_refuseæ•°"] / g[f"{lang}_æ€»æ•°"]).fillna(0.0)
        out = g if out is None else out.join(g, how="outer")
    out = out.reset_index().rename(columns={group_col: "åˆ†ç»„"})
    try:
        out = out.sort_values("åˆ†ç»„").reset_index(drop=True)
    except Exception:
        pass
    return out

# ===== Step 5. è®¡ç®—ç»“æœ =====
overall_table = overall_rates_by_language(df)
category_table = rates_by_group(df, group_col="Category")
method_table = rates_by_group(df, group_col="Rewrite Method")

# ===== Step 6. è¾“å‡ºæ±‡æ€» =====
print("===== ğŸ“Š å„è¯­è¨€æ€»ä½“å›ç­”/æ‹’ç­”ç‡ =====")
display(overall_table)

print("\n===== ğŸ“˜ æŒ‰ Category çš„ä¸‰è¯­è¨€å›ç­”/æ‹’ç­”ç‡ =====")
display(category_table)

print("\n===== ğŸ“— æŒ‰ Rewrite Method çš„ä¸‰è¯­è¨€å›ç­”/æ‹’ç­”ç‡ =====")
display(method_table)
```

```python
# Cell 5 â€” è·¨è¯­è¨€ä¸€è‡´æ€§ï¼ˆCross-lingual Consistency, CLCï¼‰â€” å®Œæ•´å¢å¼ºç‰ˆ

# 1) æ¸…æ´—ä¸‰è¯­è¨€æ ‡ç­¾ï¼Œç»Ÿä¸€ä¸ºå°å†™å­—ç¬¦ä¸²
triples = pd.DataFrame({
    "EN": df["result_en_label"].astype(str).str.lower(),
    "CN": df["result_cn_label"].astype(str).str.lower(),
    "MIX": df["result_mix_label"].astype(str).str.lower(),
})

# 2) ä¸‰ç±»ä¸€è‡´æ€§ç±»å‹çš„å¸ƒå°”æ©ç 
nuniq = triples.nunique(axis=1)
all_equal = (nuniq == 1)      # ä¸‰è€…ä¸€è‡´ï¼šEN==CN==MIX
two_equal = (nuniq == 2)      # ä¸¤ä¸¤ä¸€è‡´ï¼šä¸‰è€…ä¸­æœ‰ä¸¤ä¸ªç›¸åŒã€ä¸€ä¸ªä¸åŒ
all_diff  = (nuniq == 3)      # å®Œå…¨ä¸ä¸€è‡´ï¼šä¸‰è€…éƒ½ä¸åŒï¼ˆanswer/refuseä¸¤ç±»ä¸‹é€šå¸¸ä¸º0ï¼‰

# 3) æ€»ä½“ä¸€è‡´æ€§ç»Ÿè®¡è¡¨
stats = pd.DataFrame({
    "ç±»å‹": ["ä¸‰è€…ä¸€è‡´", "ä¸¤ä¸¤ä¸€è‡´", "å®Œå…¨ä¸ä¸€è‡´"],
    "æ•°é‡": [all_equal.sum(), two_equal.sum(), all_diff.sum()],
    "å æ¯”": [all_equal.mean(), two_equal.mean(), all_diff.mean()]
})
display(stats)

# 4) ç»„åˆæ¨¡å¼åˆ†å¸ƒï¼ˆä¾‹å¦‚ "answer-answer-refuse"ï¼‰
pattern_series = (triples["EN"] + "-" + triples["CN"] + "-" + triples["MIX"])
pattern_table = pattern_series.value_counts(dropna=False)\
                              .rename_axis("ç»„åˆæ¨¡å¼")\
                              .reset_index(name="æ•°é‡")
pattern_table["å æ¯”"] = pattern_table["æ•°é‡"] / len(triples) if len(triples) else 0.0
display(pattern_table)

# 5) â€œä¸¤ä¸¤ä¸€è‡´â€æ—¶çš„â€œå¼‚ç±»è¯­è¨€â€å½’å› åˆ†æï¼ˆè°ä¸å¦å¤–ä¸¤è€…ä¸åŒï¼‰
def identify_odd_language(row):
    en, cn, mix_ = row["EN"], row["CN"], row["MIX"]
    # åªæœ‰åœ¨ä¸¤ä¸¤ä¸€è‡´çš„æƒ…å†µä¸‹æ‰åˆ¤æ–­â€œå¼‚ç±»è¯­è¨€â€
    if (len({en, cn, mix_}) == 2):
        if cn == mix_ and en != cn:
            return "EN"    # EN æ˜¯å¼‚ç±»ï¼ŒCN==MIX
        if en == mix_ and cn != en:
            return "CN"    # CN æ˜¯å¼‚ç±»ï¼ŒEN==MIX
        if en == cn and mix_ != en:
            return "MIX"   # MIX æ˜¯å¼‚ç±»ï¼ŒEN==CN
    return "N/A"

odd_lang = triples[two_equal].apply(identify_odd_language, axis=1)
odd_table = odd_lang.value_counts(dropna=False).rename_axis("å¼‚ç±»è¯­è¨€")\
                    .reset_index(name="æ•°é‡")
if two_equal.sum() > 0:
    odd_table["å æ¯”ï¼ˆç›¸å¯¹äºä¸¤ä¸¤ä¸€è‡´æ ·æœ¬ï¼‰"] = odd_table["æ•°é‡"] / two_equal.sum()
else:
    odd_table["å æ¯”ï¼ˆç›¸å¯¹äºä¸¤ä¸¤ä¸€è‡´æ ·æœ¬ï¼‰"] = 0.0

display(odd_table)

# 6) ï¼ˆå¯é€‰ï¼‰è¡¥å……ï¼šåˆ†åˆ«è¾“å‡ºâ€œä¸‰è€…ä¸€è‡´æ ·æœ¬â€çš„æ¨¡å¼åˆ†å¸ƒä¸â€œä¸¤ä¸¤ä¸€è‡´æ ·æœ¬â€çš„æ¨¡å¼åˆ†å¸ƒ
# ä¾¿äºæ›´ç»†ç²’åº¦è§‚å¯Ÿä¸€è‡´/ä¸ä¸€è‡´æ—¶çš„æ ‡ç­¾ç»„åˆ
pattern_equal = pattern_series[all_equal].value_counts().rename_axis("ä¸‰è€…ä¸€è‡´_æ¨¡å¼").reset_index(name="æ•°é‡")
pattern_twoeq  = pattern_series[two_equal].value_counts().rename_axis("ä¸¤ä¸¤ä¸€è‡´_æ¨¡å¼").reset_index(name="æ•°é‡")
display(pattern_equal)
display(pattern_twoeq)
```
