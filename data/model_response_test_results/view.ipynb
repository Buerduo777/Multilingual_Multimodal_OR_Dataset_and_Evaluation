```python
import pandas as pd
import numpy as np

# å‡è®¾æ‚¨çš„ CSV æ–‡ä»¶åœ¨è¿™é‡ŒåŠ è½½æˆåŠŸ
# è¯·ç¡®ä¿ "test_gemini25flash_on_local_data_results.csv" æ–‡ä»¶ä¸æ‚¨çš„ Python è„šæœ¬åœ¨åŒä¸€ç›®å½•ä¸‹
data = pd.read_csv("test_deepseekv32_on_local_data_results.csv")

df = pd.DataFrame(data)

def check_missing_values(df):
    """
    æ£€æŸ¥ Pandas DataFrame ä¸­çš„ç©ºå€¼ï¼Œå¹¶è¿”å›æ¯åˆ—çš„ç©ºå€¼æ€»æ•°å’Œç™¾åˆ†æ¯”ã€‚
    """
    # 1. ä½¿ç”¨ isnull() æ£€æŸ¥ç©ºå€¼ï¼Œå¹¶ç”¨ sum() ç»Ÿè®¡æ¯åˆ—çš„ç©ºå€¼æ•°é‡
    missing_count = df.isnull().sum()

    # 2. è¿‡æ»¤æ‰ç©ºå€¼æ•°é‡ä¸º 0 çš„åˆ—ï¼Œåªä¿ç•™æœ‰ç©ºå€¼çš„åˆ—
    missing_info = missing_count[missing_count > 0]

    # 3. è®¡ç®—ç©ºå€¼ç™¾åˆ†æ¯”
    missing_percentage = (missing_info / len(df)) * 100

    # 4. ç»„åˆç»“æœå¹¶æ ¼å¼åŒ–è¾“å‡º
    missing_data_report = pd.DataFrame({
        'Missing Count': missing_info,
        'Missing Percentage': missing_percentage.round(2).astype(str) + '%'
    })

    # 5. æ ¹æ®ç©ºå€¼æ•°é‡é™åºæ’åº
    missing_data_report = missing_data_report.sort_values(by='Missing Count', ascending=False)

    if missing_data_report.empty:
        print("ğŸ‰ æ•°æ®é›†ä¸­æ²¡æœ‰æ£€æµ‹åˆ°ä»»ä½•ç©ºå€¼ã€‚")
        return None
    else:
        print("âš ï¸ æ•°æ®é›†ä¸­æ£€æµ‹åˆ°ç©ºå€¼ (NaN) çš„åˆ—ä¿¡æ¯:")
        return missing_data_report

def get_uids_of_null_rows(df, uid_column='Uid'):
    """
    æ‰¾å‡º DataFrame ä¸­ä»»ä½•åˆ—åŒ…å«ç©ºå€¼çš„è¡Œï¼Œå¹¶è¿”å›è¿™äº›è¡Œçš„æŒ‡å®š Uid åˆ—çš„å€¼åˆ—è¡¨ã€‚
    """
    # 1. df.isnull().any(axis=1) æ‰¾å‡ºåŒ…å«è‡³å°‘ä¸€ä¸ªç©ºå€¼çš„è¡Œï¼ˆè¿”å›ä¸€ä¸ªå¸ƒå°” Seriesï¼‰
    rows_with_nulls = df[df.isnull().any(axis=1)]

    if rows_with_nulls.empty:
        return []

    # 2. æå–è¿™äº›è¡Œçš„ 'Uid'
    uids_with_nulls = rows_with_nulls[uid_column]

    return uids_with_nulls.tolist()


# --- ä¸»æ‰§è¡Œæµç¨‹ ---

# 1. è¿è¡Œæ£€æŸ¥å‡½æ•°å¹¶æ‰“å°ç©ºå€¼æŠ¥å‘Š
report = check_missing_values(df)

if report is not None:
    print("\n--- ç©ºå€¼æŠ¥å‘Š ---")
    print(report)

    # 2. å¦‚æœå­˜åœ¨ç©ºå€¼ï¼Œåˆ™æå–å¹¶æ‰“å°åŒ…å«ç©ºå€¼çš„è¡Œçš„ Uid
    uids_list = get_uids_of_null_rows(df, uid_column='Uid')

    if uids_list:
        print("\n--- åŒ…å«ç©ºå€¼çš„è¡Œçš„ Uid åˆ—è¡¨ ---")
        print(f"æ€»è®¡ {len(uids_list)} è¡ŒåŒ…å«ç©ºå€¼ã€‚")
        # ä¸ºäº†é¿å…åˆ—è¡¨è¿‡é•¿ï¼Œåªæ‰“å°å‰10ä¸ª Uid å’Œæ€»æ•°
        if len(uids_list) > 10:
             print(f"å‰ 10 ä¸ª Uid: {uids_list[:10]} ... (è¿˜æœ‰ {len(uids_list) - 10} ä¸ª)")
        else:
             print(uids_list)

else:
    # å¦‚æœ report ä¸º Noneï¼Œè¯´æ˜æ²¡æœ‰ç©ºå€¼ï¼Œä¸éœ€è¦æ‰§è¡Œ Uid æå–
    pass


# 3. é¢å¤–æ£€æŸ¥ï¼šæ£€æŸ¥æ•´ä¸ª DataFrame æ˜¯å¦æœ‰ä»»ä½•ç©ºå€¼
if df.isnull().values.any():
    print("\nğŸ’¡ æ•´ä¸ª DataFrame ä¸­å­˜åœ¨è‡³å°‘ä¸€ä¸ªç©ºå€¼ã€‚")
else:
    print("\nğŸ’¡ æ•´ä¸ª DataFrame ä¸­ä¸å­˜åœ¨ä»»ä½•ç©ºå€¼ã€‚")
```
